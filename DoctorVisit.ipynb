```python
# Distribution of Visits Across Different Age Groups
import plotly.express as px
import pandas as pd

# Creating age bins (same as in your code)
age_bins = [0, 0.3, 0.5, 0.7, 1.0]  # Define age group ranges (since age is divided by 100)
age_labels = ['0-30', '30-50', '50-70', '70+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Interactive box plot for distribution of visits by age group and gender
fig = px.box(df, 
             x="age_group", 
             y="visits", 
             color="gender", 
             title="Distribution of Doctor Visits Across Different Age Groups", 
             color_discrete_sequence=px.colors.qualitative.Set2,  # Use colorful palette
             labels={'visits': 'Number of Visits', 'age_group': 'Age Group'},
             points="all"  # Show all data points as well for better insight
            )

# Customize layout to make the plot more appealing
fig.update_layout(
    xaxis_title="Age Group",
    yaxis_title="Number of Visits",
    plot_bgcolor='rgba(240, 240, 240, 0.9)',  # Light background for better contrast
    hoverlabel=dict(bgcolor='white', font_size=13, font_family="Arial"),  # Hover label styling
    template='plotly',  # Interactive plotly theme
)

# Show the interactive plot
fig.show()
```

```python
# Number of Days of Reduced Activity by Gender (Male vs Female)
import plotly.express as px

# Grouping by gender to find total reduced activity days
reduced_activity_gender = df.groupby('gender')['reduced'].sum().reset_index()

# Bar plot to visualize reduced activity by gender
fig = px.bar(reduced_activity_gender, 
             x='gender', 
             y='reduced', 
             title="Total Reduced Activity Days by Gender",
             labels={'reduced': 'Total Reduced Activity Days', 'gender': 'Gender'},
             color='gender', 
             color_discrete_sequence=px.colors.qualitative.Set2)

fig.show()
```

```python
# How Income Affects the Number of Visits to the Hospital 
import plotly.express as px

# Scatter plot to visualize income vs visits
fig = px.scatter(df, 
                 x='income', 
                 y='visits', 
                 color='gender', 
                 title="Income vs Number of Doctor Visits",
                 labels={'income': 'Annual Income (in tens of thousands)', 'visits': 'Number of Visits'})

fig.show()
```

```python
# Count the number of people with each health insurance type
insurance_types = df[['freepoor', 'freerepat', 'private']].apply(pd.Series.value_counts).T

# Print the structure of insurance_types to check how it looks
print(insurance_types)
```

```python
# Calculate the percentage of people with each type of insurance
insurance_types['percentage'] = insurance_types['yes'] / len(df) * 100

# Pie plot for the percentage of people with each type of insurance
fig = px.pie(insurance_types, 
             names=insurance_types.index, 
             values='percentage', 
             title="Percentage of People with Different Types of Health Insurance")

fig.show()
```

```python
import plotly.express as px

# Count of males and females affected by illness
illness_gender = df[df['illness'] > 0].groupby('gender').size().reset_index(name='count')

# Donut chart to show the number of males and females affected by illness
fig = px.pie(illness_gender, 
             names='gender', 
             values='count', 
             title="Number of Males and Females Affected by Illness", 
             hole=0.4,  # Creating a donut chart
             color='gender', 
             color_discrete_sequence=['#66b3ff', '#ff6666'], 
             labels={'count': 'Number of Affected People'})

# Update layout for better appearance
fig.update_layout(
    annotations=[dict(text='Affected by Illness', x=0.5, y=0.5, font_size=20, showarrow=False)],
    showlegend=True
)

fig.show()
```

```python
# How does the chronic condition status in nchronic (number of chronic conditions) compare to the lchronic (long-term chronic conditions) status in terms of the number of doctor visits?
import plotly.express as px

# Create a new column for chronic condition comparison
df['chronic_condition_status'] = df.apply(lambda row: f"{row['nchronic']} Chronic vs {row['lchronic']} Long-Term", axis=1)

# Grouping by the chronic condition status and calculating average visits
chronic_visits = df.groupby('chronic_condition_status')['visits'].mean().reset_index()

# Plotting the comparison using bar chart
fig = px.bar(chronic_visits, 
             x='chronic_condition_status', 
             y='visits', 
             color='chronic_condition_status',
             title="Average Doctor Visits for Different Chronic and Long-Term Condition Combinations",
             labels={'visits': 'Average Number of Visits', 'chronic_condition_status': 'Condition Status'},
             color_discrete_sequence=['#ffcc99', '#66b3ff', '#ff6666', '#99ff99'])

fig.update_layout(
    xaxis_title='Chronic Condition Status',
    yaxis_title='Average Number of Visits',
    showlegend=False,
    xaxis_tickangle=-45,  # Rotate x-axis labels for better visibility
)

fig.show()
```

```python
''' How do the number of doctor visits correlate with factors such as age group, gender,
chronic conditions (both nchronic and lchronic), income level, and insurance status, 
and what insights can we derive about the impact of these variables on healthcare utilization? '''


import plotly.express as px
import pandas as pd
import numpy as np

# Assuming 'df' is the dataframe with relevant data

# Encode Chronic Condition as a numeric value for 3D plotting
df['chronic_condition_numeric'] = np.where(df['nchronic'] == 'Yes', 1, 0) + np.where(df['lchronic'] == 'Yes', 1, 0)

# Encode Age Group for size (numeric value)
age_mapping = {'0-30': 1, '30-50': 2, '50-70': 3, '70+': 4}
df['age_group_numeric'] = df['age_group'].map(age_mapping)

# Create a 3D scatter plot with increased width and height
fig = px.scatter_3d(df, 
                    x='visits', 
                    y='income', 
                    z='chronic_condition_numeric',
                    color='gender',           # Color by gender
                    size='age_group_numeric', # Size by age group
                    hover_data=['age_group', 'nchronic', 'lchronic'], # Hover data for more context
                    title="Doctor Visits, Income, and Chronic Condition Status",
                    labels={'visits': 'Number of Visits', 
                            'income': 'Income (in ten thousand)', 
                            'chronic_condition_numeric': 'Chronic Condition Status', 
                            'gender': 'Gender', 
                            'age_group_numeric': 'Age Group'},
                    opacity=0.7,
                    width=1200,               # Increase the width
                    height=800)               # Increase the height

# Show the plot
fig.show()
```
